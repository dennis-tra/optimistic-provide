```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
```

```python
ROUNDS = 10_000
N = 17_000
k = 20

closest = []
second_closest = []
third_closest = []
for i in range(ROUNDS):

    peer_locations = np.random.uniform(0, 1, size=N)
    cid_location = np.random.uniform(0, 1)

    sorted_distances = np.sort(np.abs(peer_locations - cid_location))
    
    closest += [sorted_distances[0]]
    second_closest += [sorted_distances[1]]
    third_closest += [sorted_distances[2]]
```

```python
fig, ax = plt.subplots(figsize=[20, 8])

ax.hist(closest, bins=100, alpha=0.6)
ax.hist(second_closest, bins=100, alpha=0.6)
ax.hist(third_closest, bins=100, alpha=0.6)
ax.axvline(np.mean(closest))
ax.axvline(np.mean(second_closest))
ax.axvline(np.mean(third_closest))
```
